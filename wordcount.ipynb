```python
import pyspark 
from pyspark.context import SparkContext
print(pyspark.__version__)
```

```python
sc=SparkContext("local","PySpark Word Count Example")
words= sc.textFile('./file.txt').flatMap(lambda line : line.split(""))
wordCounts = words.map(lambda wird:(word , 1)).reduceByKey(lambda a, b : a+b)
wordCounts.saveAsTextFile("./TEL3")



```
